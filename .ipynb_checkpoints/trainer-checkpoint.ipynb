```python
import os
import cv2
import numpy as np
from PIL import Image
```

```python
recognizer=cv2.face.LBPHFaceRecognizer_create() 
```

```python
path='dataset'
```

```python
def get_images_with_id(path):
    images_paths=[os.path.join(path,f) for f in os.listdir(path)] #set image path to os
    faces=[]
    ids=[]
    for single_image_path in images_paths:
        faceImg=Image.open(single_image_path).convert('L')#image converted into gray color 
        faceNp=np.array(faceImg,np.uint8)
        id=int(os.path.split(single_image_path)[-1].split(".")[1])
        print(id)
        faces.append(faceNp)
        ids.append(id)
        cv2.imshow("Training",faceNp)
        cv2.waitKey(10)
    return np.array(ids),faces
```

```python
ids,faces=get_images_with_id(path)
```

```python
recognizer.train(faces,ids)
```

```python
recognizer.save('Recognizer/trainingdata.yml')
```

```python
cv2.destroyAllWindows()
```
