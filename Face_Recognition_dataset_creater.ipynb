```python
import cv2
import numpy as np
import sqlite3 as sql
```

```python
def insertupdate(id,name,age):#function for sqlite database
    conn=sql.connect('sqlite.db')#connect to database
    cmd='Select * from Students where id='+str(id)
    cursor=conn.execute(cmd) #cursor to execute statement
    isRecordexist=0#assume there is no record in our table
    for row in cursor:
        isRecordexist=1
    if (isRecordexist==1):#if there exists a record in our table
        conn.execute('UPDATE Students SET Name=? Where Id=?',(Name,Id,))
        conn.execute('UPDATE Students SET Age=? Where Id=?',(Age,Id,))
    else:#if the record does not exists
        conn.execute('INSERT into Students (Id,Name,Age) VALUES(?,?,?)',(Id,Name,Age))
    conn.commit()
    conn.close()
        
```

```python
face_detect=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
video_cap=cv2.VideoCapture(0)
```

```python
Id=input('Enter user id: ')
Name=input('Enter user name: ')
Age=input('Enter user age: ')
```

```python
insertupdate(Id,Name,Age)
```

```python
#detect face in web camera coding
sampleNum=0#assuming dataset is empty
while(True):
    ret,video_data=video_cap.read()
    gray=cv2.cvtColor(video_data,cv2.COLOR_BGR2GRAY) #Image Convert into BGRGRAY Color
    faces=face_detect.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5)
    for (x,y,w,h) in faces:
        sampleNum=sampleNum+1 #if face is detected increments
        cv2.imwrite('dataset/user.'+str(Id)+"."+str(sampleNum)+".jpg",gray[y:y+h,x:x+w])
        cv2.rectangle(video_data,(x,y),(x+w,y+h),(0,255,0),2)
        cv2.waitKey(100) #delay time
    cv2.imshow('Face',video_data) #show faces detected in web camera
    cv2.waitKey(1)
    if(sampleNum>20):
        break
video_cap.release()
cv2.destroyAllWindows()
        
```
